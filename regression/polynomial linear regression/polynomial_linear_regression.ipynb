## Polynomial linear regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('position_salaries.csv')
x = df.iloc[:, 1:2].values
y = df.iloc[:, 2].values
```

```python
df.shape
```

```python
df
```

```python
linear_regression = LinearRegression()
linear_regression.fit(x, y)
```

```python
polynomial = PolynomialFeatures(degree=5)
x_polynomial = polynomial.fit_transform(x)
```

```python
x
```

```python
x_polynomial
```

```python
linear_regression_2 = LinearRegression()
linear_regression_2.fit(x_polynomial, y)
```

```python
plt.scatter(x, y, color='red')
plt.plot(x, linear_regression.predict(x), color='blue')
plt.title('linear regression')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape(len(x_grid), 1)
```

```python
plt.scatter(x, y, color='red')
plt.plot(x_grid, linear_regression_2.predict(polynomial.fit_transform(x_grid)), color='blue')
plt.title('polynomial regression')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```
